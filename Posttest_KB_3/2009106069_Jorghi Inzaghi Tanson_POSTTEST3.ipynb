#### Import Library yang dibutuhkan

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Baca dataset yang diunduh ke dalam bentuk dataframe dan tampilkan isi dataframe

```python
dFrame = pd.read_csv('imdb_video_game_rating.csv')
dFrame
```

#### Buatlah 3 visualisasi data dari dataframe di atas beserta kesimpulannya

#### 1) Visualisasi Data Grafik Batang

```python
# Sintaks berikut berfungsi untuk mendeklarasi variabel dataframe baru di mana isi dari dataframe
# tersebut adalah data tahun beserta jumlah game yang terilis pada tahun tersebut
jumlah_game = dFrame['year'].value_counts().rename_axis('year').reset_index(name='total')
```

#### Deklarasi grafik bar beserta propertinya

```python
plt.bar(jumlah_game['year'], jumlah_game['total'])
plt.title('Jumlah Game Berdasarkan Tahun Rilis', loc='right')
plt.xlabel('Tahun Rilis', loc='left')
plt.ylabel('Jumlah', loc='bottom')
plt.show()
```

#### Kesimpulan dari grafik di atas adalah:
* Sebelum tahun 1970, hampir tidak pernah ada game yang rilis. Namun, perkembangan industri game mulai meningkat. Hal ini ditandai dengan peningkatan jumlah game yang dirilis setelah tahun 1970 di mana pada tahun 1980 jumlah game yang dirilis meningkat pesat jika dibandingkan dengan tahun-tahun sebelumnya
* Rentang tahun 2000 sampai 2020 menjadi masa yang paling dinikmati oleh komunitas gamer karena banyak sekali judul-judul game yang dirilis. Dengan kata lain, komunitas tidak akan bosan dengan game yang itu-itu saja
* Tahun 2017 menjadi tahun di mana banyak sekali game yang dirilis, namun 2022 memiliki kemungkinan menjadi tahun yang paling banyak merilis game mengingat tahun tersebut belum berakhir

#### 2) Visualisasi Data Grafik Garis

```python
# Sintaks berikut berfungsi untuk mendeklarasi variabel dataframe baru di mana isi dari dataframe
# tersebut adalah data rating yang diurut berdasarkan tahun rilis game
yearRate = dFrame['rating'].groupby(dFrame['year']).mean().rename_axis('year').reset_index(name='mean')
```

#### Deklarasi grafik garis beserta propertinya

```python
plt.plot(yearRate['year'], yearRate['mean'])
plt.title('Rata-Rata Rating Game Berdasarkan Tahun', loc='right')
plt.xlabel('Tahun Rilis', loc='left')
plt.ylabel('Rata-Rata Rating', loc='bottom')
plt.show()
```

#### Kesimpulan dari grafik di atas adalah:
* Mayoritas game yang dirilis pada tiap tahunnya diterima oleh komunitas gamer mengingat banyak nilai rata-rata rating yang terletak di atas nilai 6.5
* Rentang tahun 1960 sampai 1980 terjadi penurunan rata-rata rating yang drastis. Hal ini mengindikasikan bahwa game-game yang rilis pada rentang tahun tersebut kurang diterima oleh komunitas gamer
* Mengetahui hal tersebut, para game developer mulai memperbaiki kualitas game mereka. Hal ini dapat dilihat pada rentang waktu 1980 sampai 1990. Pada rentang tahun tersebut terjadi kenaikan rata-rata rating yang drastis.
* Pada tahun 2020 hingga seterusnya menjadi tahun dengan rata-rata rating game tertinggi. Namun perlu diingat, rata-rata rating tersebut bisa saja turun mengingat tahun 2022 belum berakhir

#### 3) Visualisasi Data Grafik Pie

```python
# Sintaks berikut berfungsi untuk mendeklarasi variabel dataframe baru di mana isi dari dataframe
# tersebut adalah data votes yang diurut berdasarkan tahun rilis game
yearVote = dFrame['votes'].groupby(dFrame['year']).sum().rename_axis(
    'year').reset_index(name='sum')

# Selanjutnya mengubah seluruh label sebelum tahun 2001 menjadi 2000 ke bawah
# Hal ini bertujuan untuk memudahkan penafsiran dari grafik pie
# karena banyaknya label yang terdaftar membuat grafik tersebut
# menjadi tidak rapi
yearVote.loc[yearVote['year'] <= 2000, "year"] = "2000 ke Bawah"
yearVote = yearVote.groupby("year").sum().reset_index()
```

#### Deklarasi grafik pie beserta propertinya

```python
plt.figure(figsize=(12, 12))
plt.pie(yearVote['sum'], labels=yearVote['year'], startangle=90,
        autopct='%.1f%%')
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
plt.title('Persebaran Jumlah Voting Berdasarkan Tahun Rilis', loc='left')
plt.xlabel('Tahun Rilis')
plt.ylabel('Rata-Rata Voting')
plt.show()
```

#### Kesimpulan dari grafik di atas adalah:
* Tahun 2022 menjadi tahun dengan jumlah vote terdikit dengan nilai 0.5%. Hal ini menunjukan bahwa komunitas gamer kurang meminati game-game yang keluar pada tahun 2022. Namun hal ini bisa dibantah karena beberapa faktor, misal:
    * Masih sedikit game yang masuk ke dalam website IMDb mengingat tahun 2022 belum berakhir
    * Kurangnya kesadaran diri para gamer untuk melakukan voting di website IMDb
    * Komunitas gamer lebih mengenal website IMDb sebagai website informasi per-film-an
    * dsb.
* Tahun 2000 ke bawah terlihat menjadi tahun dengan jumlah voting terbanyak. Namun hal tersebut tidak valid karena tahun tersebut merupakan gabungan dari beberapa tahun, sehingga menghasilkan jumlah vote yang tergolong besar. Tahun 2000 ke bawah memiliki rentang antara 1945-2000
* Tahun 2007, 2009, 2010, dan 2011 termasuk ke dalam tahun dengan jumlah vote yang banyak dengan nilai >= 5%. Hal ini menunjukkan bahwa game-game yang dirilis pada rentang waktu tersebut banyak diminati oleh komunitas gamer pada masanya.
* Tahun 2013 menjadi tahun dengan jumlah vote tertinggi dengan nilai 7.5%. Hal ini mengindikasikan bahwa tahun tersebut menjadi tahun yang paling berpengaruh pada industri game dan akan selalu dikenang oleh komunitas gamer, khususnya mereka yang lahir pada tahun 2000-an
